Visualization of Variable Distributions in Dataset

```python
variables = ['Price', 'Age', 'KM', 'FuelType', 'HP', 'CC', 'Doors', 'Weight', 'AutoType', 'MetColorType']

fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(19, 10))
axes = axes.flatten()

for i, variable in enumerate(variables):
    ax = axes[i]
    if ucs[variable].dtype == 'object':
        ucs[variable].value_counts().plot(kind='bar', ax=ax)
    else:
        ucs[variable].plot.hist(ax=ax, bins=10)
    ax.set_xlabel(variable)
    ax.set_ylabel('Frequency')
    ax.set_title(f'Distribution of {variable}')

plt.tight_layout()

plt.show()
```

Checking Correlation in the dataset

```python
corelation = ucs.corr()
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns,annot=True)
```

The Age and KM have the negative correlation with Price and HP,CC,Doors and Weight have the positive correlation with Price

What is the average price of each fuel type?

```python
avg_price_by_fuel = ucs.groupby("FuelType")["Price"].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(x="FuelType", y="Price", data=avg_price_by_fuel, palette="Set2")
plt.title("Average Price by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Average Price")
plt.show()
```

Cars with Diesel fuel type has the highest avg price among the 3 fuel types

**Scatter Plot: Age vs. Price by Fuel Type**

```python
px.scatter(ucs, x='Age', y='Price', color='FuelType', hover_data=['KM','HP','CC','Doors','Weight'])
```

cars aging between 0 to 10 have the highest Price and gradually decreasing by increas in age.

How Kilometers driven affect the price of the used cars?

```python
plt.scatter(ucs['KM'], ucs['Price'], marker='1')
plt.title("Price impact based on KM driven")
plt.xlabel(["Kilometers"])
plt.ylabel(["Price"])
```

Cars having less kilometers driven have the highest price and decreasing by increase in kilometers

How Age of car impact the car price??

```python
plt.figure(figsize=(8, 6))
sns.regplot(x=ucs['Age'], y=ucs['Price'])
plt.title('Impact of Age on Car Price')
plt.xlabel('Age (Years)')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

Price is inversely propositional to Age.

What is the distribution of car ages in the dataset?

```python
plt.figure(figsize=(8, 6))
sns.histplot(ucs['Age'], bins=20, kde=True)
plt.title("Distribution of Car Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
